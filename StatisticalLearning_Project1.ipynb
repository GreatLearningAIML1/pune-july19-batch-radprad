###### Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. 

```python
# read the data in pandas FROM A CSV
import pandas as pd

df = pd.read_csv("Laliga.csv",skiprows=1,skipinitialspace=False,index_col="Team")
df.replace(to_replace='-',value=0,inplace=True)
```

###### Print all the teams which have started playing between 1930-1980.

```python
df["StartPlay"]=df['Debut'].str.slice(0,4,1).astype('int64')
df.loc[((df.StartPlay >= 1930) & (df.StartPlay <= 1980)),:]
```

###### Print the list of teams which came Top 5 in terms of points 

```python
dfnew=df.groupby('Team').Points.max().to_frame()
dfnew['Points']=dfnew['Points'].astype(int)
dfnew.sort_values(by=['Points'],ascending=False,inplace=True)
print("Top 5 teams having maximum points \n",dfnew['Points'].head(5))
```

###### Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference

```python
def Goal_diff_count(GoalsFor, GoalsAgainst):
    return (GoalsFor - GoalsAgainst)
```

```python
##Method 1 using function in single row
df['GoalsFor']=df['GoalsFor'].astype(int)
df['GoalsAgainst']=df['GoalsAgainst'].astype(int)
print("Team with maximum goal difference : ",df.apply(lambda x: Goal_diff_count(x['GoalsFor'], x['GoalsAgainst']), axis=1).idxmax())
print("Team with minimum goal difference : ",df.apply(lambda x: Goal_diff_count(x['GoalsFor'], x['GoalsAgainst']), axis=1).idxmin())
```

```python
###Method 2 long way
df['GoalsFor']=df['GoalsFor'].astype(int)
df['GoalsAgainst']=df['GoalsAgainst'].astype(int)
df['GoalDiff']=Goal_diff_count(df['GoalsFor'],df['GoalsAgainst'])
dfnew=df.groupby('Team').GoalDiff.max().to_frame()
dfnew.sort_values(['GoalDiff'],ascending=False,inplace=True)
print("Team with maximum goal difference : ",dfnew.index[0])
dfnew.sort_values(['GoalDiff'],ascending=True,inplace=True)
print("Team with minimum goal difference : ",dfnew.index[0])
```

###### Create a new column with name “Winning Percent” and append it to the data set 
Percentage of Winning = (GamesWon / GamesPlayed)*100
If there are any numerical error, replace it with 0%

```python
df['GamesWon']=df['GamesWon'].astype(int)
df['GamesPlayed']=df['GamesPlayed'].astype(int)
df['PercentageOfWinning']=(df['GamesWon'] / df['GamesPlayed']) * 100
df['PercentageOfWinning'].fillna(0,inplace=True)
```

###### Print the top 5 teams which has the highest Winning percentage

```python
df.sort_values(['PercentageOfWinning'],ascending=False,inplace=True)
df.iloc[:5,21]
```

###### Group teams based on their “Best position” and print the sum of their points for all positions (5 points) Eg: Best Position Points

      1                              25000

      2                              7000  

```python
df['BestPosition']=df['BestPosition'].astype(int)
df['Points']=df['Points'].astype(int)
ds_suml1=df.groupby(['BestPosition','Team'])["Points"].sum()
print(pd.concat([ds_suml1.to_frame(),
           ds_suml1.sum(level=0).to_frame()
                 .assign(Team= "Total")
                 .set_index('Team', append=True)])
  .sort_index())
```
